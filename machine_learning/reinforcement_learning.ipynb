# üéÆ Reinforcement Learning with Q-Learning (FrozenLake)

This notebook introduces **Q-learning**, a value-based reinforcement learning algorithm, using the `FrozenLake` environment from OpenAI Gym.

We will:
- Understand the RL setup (Agent, Environment, Rewards)
- Implement a Q-table-based learning agent
- Train the agent using Q-learning
- Evaluate the learned policy


```python
# üì¶ Install required libraries (run only once in Codespaces or Colab)
# !pip install gymnasium

import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

## üåç Step 1: Initialize FrozenLake Environment
FrozenLake is a 4x4 grid world where the agent must learn to reach a goal while avoiding holes.


```python
# Initialize FrozenLake environment (non-slippery version for faster learning)
env = gym.make("FrozenLake-v1", is_slippery=False)

state_space = env.observation_space.n
action_space = env.action_space.n

print(f"State space: {state_space}, Action space: {action_space}")
```

## üß† Step 2: Initialize Q-Table and Set Hyperparameters
We create a Q-table with dimensions `[state_space x action_space]`, initialized to zeros.

We'll also define key Q-learning parameters:
- Learning rate (alpha)
- Discount factor (gamma)
- Exploration rate (epsilon) and its decay


```python
# Initialize Q-table with zeros
q_table = np.zeros((state_space, action_space))

# Hyperparameters
alpha = 0.1       # learning rate
gamma = 0.99      # discount factor
epsilon = 1.0     # exploration rate
epsilon_min = 0.01
epsilon_decay = 0.995

# Training parameters
episodes = 1000
max_steps = 100
```

## üöÄ Step 3: Train the Agent with Q-Learning
We use the Bellman equation to iteratively update the Q-values:

**Q(s, a) ‚Üê Q(s, a) + Œ± [r + Œ≥ max(Q(s', a')) ‚àí Q(s, a)]**

The agent explores with Œµ-greedy strategy and gradually shifts to exploitation.


```python
rewards = []

for episode in range(episodes):
    state, _ = env.reset()
    total_reward = 0

    for step in range(max_steps):
        # Œµ-greedy strategy
        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()  # explore
        else:
            action = np.argmax(q_table[state, :])  # exploit

        # Take action
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # Bellman update rule
        best_next_action = np.max(q_table[next_state, :])
        q_table[state, action] = q_table[state, action] + alpha * (
            reward + gamma * best_next_action - q_table[state, action]
        )

        state = next_state
        total_reward += reward

        if done:
            break

    # Decay exploration rate
    epsilon = max(epsilon_min, epsilon * epsilon_decay)
    rewards.append(total_reward)

    if (episode + 1) % 100 == 0:
        print(f"Episode {episode+1}: Total Reward = {total_reward:.0f}, Epsilon = {epsilon:.3f}")
```

## üéØ Step 4: Evaluate the Trained Agent
After training, we let the agent act **greedily** using the learned Q-table and watch how well it performs.  
We‚Äôll also calculate the **average reward** over a number of evaluation episodes.


```python
# Evaluate agent after training
eval_episodes = 20
eval_rewards = []

for ep in range(eval_episodes):
    state, _ = env.reset()
    total_reward = 0
    print(f"\nEpisode {ep+1}")
    time.sleep(0.5)

    for step in range(max_steps):
        clear_output(wait=True)
        env.render()
        action = np.argmax(q_table[state, :])
        state, reward, terminated, truncated, _ = env.step(action)
        total_reward += reward
        time.sleep(0.2)
        if terminated or truncated:
            break

    eval_rewards.append(total_reward)
    print(f"Reward: {total_reward}")
    time.sleep(1)

env.close()
print(f"\n‚úÖ Average reward over {eval_episodes} episodes: {np.mean(eval_rewards):.2f}")
```

## üìà Step 5: Visualize Training Rewards
Let's visualize how the agent's total reward per episode evolved over time to understand learning progress.


```python
# Plot total reward per episode
plt.figure(figsize=(10, 5))
plt.plot(rewards, color='dodgerblue')
plt.title('Total Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.grid(True)
plt.show()
```
